```python
# Türkiye'nin yıllara göre ihracatı
import matplotlib.pyplot as plt
import numpy as np
yillar = np.arange(2004, 2018)
buyume = [73.476, 85.535, 107.272, 132.027, 102.143, 113.883, 134.907,152.462, 151.803, 157.610, 143.839, 142.530, 156.993, 167.921]
plt.plot(yillar, buyume)
plt.title("Türkiye'nin İhracat'ı (2005-2018)")
plt.xlabel("Yıllar")
plt.ylabel("İhracat (Milyon $)")
plt.xticks([2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.show()
```

```python
#Türkiye'nin Yıllara Göre İthalat'ı
import matplotlib.pyplot as plt
import numpy as np
yillar = np.arange(2004, 2018)
buyume = [116.774, 139.576, 170.063, 201.964, 140.928, 185.544, 240.842,
         236.545, 251.661, 242.177, 207.234, 198.618, 233.800, 223.047]
plt.plot(yillar, buyume)
plt.title("Türkiye'nin İthalat'ı (2005-2018)")
plt.xlabel("Yillar")
plt.ylabel("İthalat (Milyon $)")
plt.xticks([2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.show()
```

```python
# İthalat İhracat 
import matplotlib.pyplot as plt
import numpy as np
ihracat = [73.476, 85.535, 107.272, 132.027, 102.143, 113.883, 134.907,
          152.462, 151.803, 157.610, 143.839, 142.530, 156.993, 167.921]

ithalat = [116.774, 139.576, 170.063, 201.964, 140.928, 185.544, 240.842,
         236.545, 251.661, 242.177, 207.234, 198.618, 233.800, 223.047]

yillar = np.arange(2004, 2018)
plt.title("Türkiye'nin İthalat ve İhracat Miktarları (milyon $)")
plt.plot(yillar, ihracat, 'r--', label = 'İHRACAT')
plt.plot(yillar, ithalat, 'b:', label = 'İTHALAT')
plt.legend(loc = 'upper left')
plt.show()
```
